```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python

    
def plot_bgrid_S3_z0(n=4):
    data = np.genfromtxt("s3_z0_bgrid.txt",delimiter="\t")
    grid_n = int(data.shape[0]/n)
    list_om = [r'$\Omega_{m0}=0.1$',r'$\Omega_{m0}=0.2$',r'$\Omega_{m0}=0.3$',r'$\Omega_{m0}=0.4$']
    list_symb = ["b*-","r.-","g--","k-"]
    list_ms = [5.5 for i in range(len(list_om))]
    print("grid_n is ",grid_n)
    om = data[:,0]
    b1= data[:,1]
    s3 = data[:,2]
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3(z=0)$",fontsize=40)
    ax1.set_xlabel(r"$B_1$",fontsize=40)
   # ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for omval in list_om:
        ax1.plot(b1[i*grid_n:(i+1)*grid_n],s3[i*grid_n:(i+1)*grid_n],list_symb[i],ms=list_ms[i],label=list_om[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("bgrid_S3.pdf",dpi=fig.dpi,bbox_inches='tight')
    
    return b1,s3
```

```python

    
def plot_omgrid_S3_z0(n=4):
    data = np.genfromtxt("s3_z0_omgrid.txt",delimiter="\t")
    grid_n = int(data.shape[0]/n)
    list_b1 = [r'$B_1=0.0$',r'$B_1=0.2$',r'$B_1=0.5$',r'$B_1=1.0$']
    list_symb = ["b*-","r.-","g--","k-"]
    list_ms = [5.5 for i in range(len(list_b1))]
    print("grid_n is ",grid_n)
    om = data[:,0]
    b1= data[:,1]
    s3 = data[:,2]
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3(z=0)$",fontsize=40)
    ax1.set_xlabel(r"$\Omega_{dm0}$",fontsize=40)
   # ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for b1val in list_b1:
        ax1.plot(om[i*grid_n:(i+1)*grid_n],s3[i*grid_n:(i+1)*grid_n],list_symb[i],ms=list_ms[i],label=list_b1[i])
        i = i+1
    
    ax1.axhline(y=4.857, color='m', linestyle='-',label="EDS")
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("omgrid_S3.pdf",dpi=fig.dpi,bbox_inches='tight')
    
    return b1,s3 
```

```python
b1,s3=plot_bgrid_S3_z0()
```
