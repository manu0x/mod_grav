```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
d_ncc = np.genfromtxt("wgt_nccw_om_0.279_omL_-0.8182_w_-1.061.txt",delimiter="\t")
d_B = np.genfromtxt("wgt_B1_om_0.2791_B1_1.69038.txt",delimiter="\t")
d_l = np.genfromtxt("wgt_lc_om_0.315_B1_0.0.txt",delimiter="\t")
```

```python
def plot_wgt():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel(r"$w_{gt}(\mu K)$",fontsize=30)
    ax.set_xlabel(r"$\theta (degrees)$",fontsize=30)
    ax.set_xlim([0.1, 100.0])
    
    ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],(1e6)*d_ncc[:,1],"r-",ms=1.0,label="cWNCC DE")
    ax.plot(d_B[:,0],(1e6)*d_B[:,1],"g--",ms=1.0,label="Bimetric MG")
    ax.plot(d_l[:,0],(1e6)*d_l[:,1],"b-.",ms=1.0,label="$\Lambda CDM$")
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("wgt.pdf",dpi=fig.dpi,bbox_inches='tight')



def plot_wgt_rel():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel(r"$\frac{w_{gt}(\mu K)}{^{\Lambda CDM}w_{gt}}$",fontsize=30)
    ax.set_xlabel(r"$\theta (degrees)$",fontsize=30)
    ax.set_xlim([0.1, 100.0])
    
    ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],d_ncc[:,1]/d_l[:,1],"r-",ms=1.0,label="cWNCC DE")
    ax.plot(d_B[:,0],d_B[:,1]/d_l[:,1],"b--",ms=1.0,label="Bimetric MG")
    #ax.plot(d_B[:,0],d_B[:,1],"g--",ms=1.0,label="Bimetric MG")
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("wgt_rel.pdf",dpi=fig.dpi,bbox_inches='tight')

```

```python
plot_wgt()
```

```python
plot_wgt_rel()
```
