```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
d_ncc = np.genfromtxt("wgt_nccw_om_0.279_omL_-0.8182_w_-1.061.txt",delimiter="\t")
d_B = np.genfromtxt("wgt_B1_om_0.2791_B1_1.69038.txt",delimiter="\t")
d_l = np.genfromtxt("wgt_LC_om_0.3_B1_0.0.txt",delimiter="\t")
n = d_ncc.shape[0]
print(n)
for i in range(n):
    d_ncc[i,1] = 1000000.0*d_ncc[i,1];
    d_B[i,1] = 1000000.0*d_B[i,1];
    d_l[i,1] = 1000000.0*d_l[i,1];
```

```python
def plot_wgt():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel("$w_{gT}(\mu K)$",fontsize=30)
    ax.set_xlabel("$k(hMpc^{-1})$",fontsize=30)
    ax.set_xlim([0.1, 100.0])
    
    ax.set_xscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],d_ncc[:,1],"r-",ms=1.0,label="cWNCC")
    #ax.plot(d_B[:,0],d_B[:,1],"g--",ms=1.0,label="Bi-metric")
    ax.plot(d_l[:,0],d_l[:,1],"b-.",ms=1.0,label="$\Lambda CDM$")
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("w.pdf",dpi=fig.dpi,bbox_inches='tight')
    

def plot_wgt_rel():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel("$\Delta w_{gT}(\%)$",fontsize=30)
    ax.set_xlabel("$k(hMpc^{-1})$",fontsize=30)
    ax.set_xlim([0.1, 100.0])
    
    ax.set_xscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],100.0*(d_ncc[:,1]-d_l[:,1]),"r-",ms=1.0,label="cWNCC")
    #ax.plot(d_B[:,0],(d_B[:,1]/d_l[:,1])*100.0,"g--",ms=1.0,label="Bi-metric")
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("wgt_comp.pdf",dpi=fig.dpi,bbox_inches='tight')
```

```python
plot_wgt()
```

```python
plot_wgt_rel()
```
