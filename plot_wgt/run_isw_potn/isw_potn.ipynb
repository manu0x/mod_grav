```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
d_ncc = np.genfromtxt("delta_nccw_om_0.279_omL_-0.8182_w_-1.061.txt",delimiter="\t")
d_B = np.genfromtxt("delta_B1_om_0.2791_B1_1.69038.txt",delimiter="\t")
d_l = np.genfromtxt("delta_LC_om_0.315_B1_0.0.txt",delimiter="\t")
```

```python
def plot_HbyH0():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel(r"$\frac{H}{H_0}$",fontsize=30)
    ax.set_xlabel("$a$",fontsize=30)
    ax.set_xlim([0.01, 1.0])
    
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],d_ncc[:,-2],"r-",ms=1.0,label="cWNCC DE")
    ax.plot(d_B[:,0],d_B[:,-2],"g--",ms=1.0,label="Bimetric MG")
    ax.plot(d_l[:,0],d_l[:,-2],"b-.",ms=1.0,label="$\Lambda CDM$")
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("HbyH0.pdf",dpi=fig.dpi,bbox_inches='tight')



def plot_isw_potn():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel("$\dot{\Psi}-\dot{\Phi}$",fontsize=30)
    ax.set_xlabel("$a$",fontsize=30)
    ax.set_xlim([0.01, 1.0])
    
    ax.set_xscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],d_ncc[:,-1],"r-",ms=1.0,label="cWNCC DE")
    ax.plot(d_B[:,0],d_B[:,-1],"g--",ms=1.0,label="Bimetric MG")
    ax.plot(d_l[:,0],d_l[:,-1],"b-.",ms=1.0,label="$\Lambda CDM$")
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("isw_potn.pdf",dpi=fig.dpi,bbox_inches='tight')
    

def plot_isw_potn_rel():
    fig = plt.figure(figsize=(12,10),dpi=80)
    gs = gridspec.GridSpec(1,1)
    ax = plt.subplot(gs[0,0])
    ax.set_ylabel(r"$\frac{(\dot{\Psi}-\dot{\Phi})_{MG}}{(\dot{\Psi}-\dot{\Phi})_{DE}}$",fontsize=30)
    ax.set_xlabel("$a$",fontsize=30)
    ax.set_xlim([0.1, 1.0])
    
    ax.set_xscale("log")
    ax.tick_params(axis='both', which='major',labelsize=13)
    
    ax.plot(d_ncc[:,0],d_ncc[:,-1]/d_l[:,-1],"r-",ms=1.0,label="cWNCC DE")
    ax.plot(d_ncc[:,0],d_B[:,-1]/d_l[:,-1],"g--",ms=1.0,label=r"$Bimetric MG$")
    
    
    
    ax.legend(loc='best',frameon=False, fontsize=20)
    fig.savefig("isw_potn_comp.pdf",dpi=fig.dpi,bbox_inches='tight')
```

```python
plot_isw_potn()
```

```python
plot_isw_potn_rel()
```

```python
plot_HbyH0()
```
