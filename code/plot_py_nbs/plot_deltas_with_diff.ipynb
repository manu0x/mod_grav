```python
def plot_linear(c =4):
    fig=plt.figure(figsize=(12.0,15.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(2, 1,hspace=0.1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$\frac{\delta_1}{a}$",fontsize=40)
    #ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    ax2 = plt.subplot(gs[1, 0])
    ax2.set_ylabel(r"$\Delta\left(\frac{\delta_1}{a}\right)$ %",fontsize=40)
    ax2.set_xlabel("$a$",fontsize=40)
    ax2.set_xscale('log')
    ax2.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        if i>0:
            ax2.plot(mod[:,0],100.0*np.abs(mod[:,c]-list_mod[0][:,c])/list_mod[0][:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)
    ax2.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("linear.pdf",dpi=fig.dpi,bbox_inches='tight')
    
    
def plot_2nd(c =7):
    fig=plt.figure(figsize=(12.0,15.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$\frac{\delta_2}{{\delta_2}_{ini}}$",fontsize=40)
    #ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.set_yscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    ax2 = plt.subplot(gs[1, 0])
    ax2.set_ylabel(r"$\Delta\left(\frac{\delta_2}{{\delta_2}_{ini}}\right)$ %",fontsize=40)
    ax2.set_xlabel("$a$",fontsize=40)
    ax2.set_xscale('log')
    ax1.set_yscale('log')
    ax2.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_mod:
        ax1.plot(mod[:,0],mod[:,c]/mod[0,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        if i>0:
            ax2.plot(mod[:,0],100.0*np.abs(mod[:,c]-list_mod[0][:,c])/list_mod[0][:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)
    ax2.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("second_ord.pdf",dpi=fig.dpi,bbox_inches='tight')

def plot_S3(c=9):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3$",fontsize=40)
    ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("S3.pdf",dpi=fig.dpi,bbox_inches='tight')
    
def plot_fs8(c=1):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel("$f\sigma_8$",fontsize=40)
    ax1.set_xlabel("$z$",fontsize=40)
    #ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    
    i=0
    for mod in list_mod_fs8:
        #s = np.argsort(mod[:,0])
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit_fs8[i])
        i = i+1
    ax1.errorbar(fs_8_data[:,0],fs_8_data[:,1],yerr =np.stack((fs_8_data[:,2],fs_8_data[:,3]),axis=0), fmt='*',label = "data fs8")
    
    ax1.legend(loc='upper right',frameon=False, fontsize=20)

    fig.savefig("fs8.pdf",dpi=fig.dpi,bbox_inches='tight')
    
def plot_lcdm_S3(c=9):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3$",fontsize=40)
    ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_lcdm_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_lcdm_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("lcdm_S3.pdf",dpi=fig.dpi,bbox_inches='tight')
```
