```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
l = np.genfromtxt("delta_lcdm_om_0.3.txt",delimiter="\t")
l_0p4 = np.genfromtxt("delta_lcdm_om_0.4.txt",delimiter="\t")
l_0p7 = np.genfromtxt("delta_lcdm_om_0.7.txt",delimiter="\t")
l_1 = np.genfromtxt("delta_lcdm_om_1.0.txt",delimiter="\t")

B1_0 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.0.txt",delimiter="\t")
B1_0p5 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.5.txt",delimiter="\t")
B1_1 = np.genfromtxt("delta_bimetric_om_0.3_B1_1.0.txt",delimiter="\t")
B1_3 = np.genfromtxt("delta_bimetric_om_0.3_B1_3.0.txt",delimiter="\t")
B1_6 = np.genfromtxt("delta_bimetric_om_0.3_B1_6.0.txt",delimiter="\t")


B1_0_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.0_fs8.txt",delimiter="\t")
B1_0p5_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.5_fs8.txt",delimiter="\t")
B1_1_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_1.0_fs8.txt",delimiter="\t")
B1_3_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_3.0_fs8.txt",delimiter="\t")
B1_6_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_6.0_fs8.txt",delimiter="\t")

fs_8_data = np.genfromtxt("data.txt",delimiter="\t")

list_mod_fs8 = [B1_0_fs8,B1_0p5_fs8,B1_1_fs8,B1_3_fs8,B1_6_fs8]
list_lcdm_mod = [l,l_0p4,l_0p7,l_1]
list_symb = ["b--","r-","g--","k--","c.","m--"]
list_ms = ["0.5","0.5","0.5","0.5","0.5","0.5"]
list_tit = [r"$\Lambda CDM$","B1=0","B1=0.5","B1=1","B1=3","B1=6"]
list_tit_fs8 = ["B1=0","B1=0.5","B1=1","B1=3","B1=6"]

list_lcdm_tit = [r"$\Lambda CDM \quad (\Omega_m=0.3)$",r"$\Omega_m=0.4$",r"$\Omega_m=0.7$",r"$\Omega_m=1.0$",]
```

```python
B1_0_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.0_fs8.txt",delimiter="\t")
B1_0p5_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_0.5_fs8.txt",delimiter="\t")
B1_1_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_1.0_fs8.txt",delimiter="\t")
B1_3_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_3.0_fs8.txt",delimiter="\t")
B1_6_fs8 = np.genfromtxt("delta_bimetric_om_0.3_B1_6.0_fs8.txt",delimiter="\t")

fs_8_data = np.genfromtxt("data.txt",delimiter="\t")
list_symb = ["b--","r-","g--","k--","m--","c."]
list_ms = ["0.5","0.5","0.5","0.5","0.5","0.5"]
list_mod_fs8 = [B1_0_fs8,B1_0p5_fs8,B1_1_fs8,B1_3_fs8,B1_6_fs8]
list_tit_fs8 = ["B1=0","B1=0.5","B1=1","B1=3","B1=6"]
```

```python
fs_8_data.shape
```

```python
def plot_linear(c = 4):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$\frac{\delta_1}{a}$",fontsize=40)
    ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("linear.eps",dpi=fig.dpi,bbox_inches='tight')

def plot_S3(c=10):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3$",fontsize=40)
    ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("S3.eps",dpi=fig.dpi,bbox_inches='tight')
    
def plot_fs8(c=1):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel("$f\sigma_8$",fontsize=40)
    ax1.set_xlabel("$z$",fontsize=40)
    #ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    
    i=0
    for mod in list_mod_fs8:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_tit_fs8[i])
        i = i+1
    ax1.errorbar(fs_8_data[:,0],fs_8_data[:,1],yerr =fs_8_data[:,2] ,fmt='*',label = "data fs8")
    
    ax1.legend(loc='upper right',frameon=False, fontsize=20)

    fig.savefig("fs8.pdf",dpi=fig.dpi,bbox_inches='tight')
    
def plot_lcdm_S3(c=9):
    fig=plt.figure(figsize=(12.0,9.5), dpi=80)
    #fig.suptitle('Overdensity at turn around', fontsize=14, fontweight='bold')
    gs = gridspec.GridSpec(1, 1)
    ax1 = plt.subplot(gs[0, 0])
    ax1.set_ylabel(r"$S_3$",fontsize=40)
    ax1.set_xlabel("$a$",fontsize=40)
    ax1.set_xscale('log')
    ax1.tick_params(axis='both', which='major',labelsize=13)
    
    i=0
    for mod in list_lcdm_mod:
        ax1.plot(mod[:,0],mod[:,c],list_symb[i],ms=list_ms[i],label=list_lcdm_tit[i])
        i = i+1
    
    
    ax1.legend(loc='best',frameon=False, fontsize=20)

    fig.savefig("lcdm_S3.eps",dpi=fig.dpi,bbox_inches='tight')
```
